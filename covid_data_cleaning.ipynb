```python
import pandas as pd

covid_df = pd.read_csv("us-counties.csv")
covid_df.head()
```

```python
covid_df.loc[covid_df["state"]=="Louisiana", "county"] = covid_df.loc[covid_df["state"]=="Louisiana"]["county"] + " Parish"
covid_df[covid_df["state"]=="Louisiana"]

covid_df.loc[~covid_df["county"].str.contains("County|Parish|Area|Municipality|Borough", regex=True), "county"] = covid_df[~covid_df["county"].str.contains("County|Parish|Area|Municipality|Borough", regex=True)]["county"] + " County"
```

```python
covid_df[covid_df["county"].str.contains("Skagway", regex=True)]
```

```python
covid_df["county"] = covid_df["county"].str.replace("Anchorage County", "Anchorage Municipality")
```

```python
covid_df[covid_df["state"].str.contains("Alaska", regex=True)]
```

```python
covid_df = covid_df[covid_df["county"].str.contains("Unknown")==False]
```

```python
covid_df[covid_df["county"].str.contains("Unknown County", regex=True)]
```

```python
covid_df[covid_df["county"].str.contains("City", regex=True)]
```

```python
covid_df["county"] = covid_df["county"].str.replace("New York City County", "New York County")
covid_df["county"] = covid_df["county"].str.replace("City County", "City")
```

```python
covid_df["county"] = covid_df["county"].str.replace("city County", "City")
```

```python
covid_df[covid_df["county"].str.contains("Petersburg", regex=True)]
```

```python
us_state_abbrev = {
'Alabama': 'AL', 'Alaska': 'AK', 'Arizona': 'AZ', 'Arkansas': 'AR', 'California': 'CA', 'Colorado': 'CO',
'Connecticut': 'CT', 'Delaware': 'DE', 'Florida': 'FL', 'Georgia': 'GA', 'Hawaii': 'HI', 'Idaho': 'ID',
'Illinois': 'IL', 'Indiana': 'IN', 'Iowa': 'IA', 'Kansas': 'KS', 'Kentucky': 'KY', 'Louisiana': 'LA',
'Maine': 'ME', 'Maryland': 'MD', 'Massachusetts': 'MA', 'Michigan': 'MI', 'Minnesota': 'MN', 'Mississippi': 'MS',
'Missouri': 'MO', 'Montana': 'MT', 'Nebraska': 'NE', 'Nevada': 'NV', 'New Hampshire': 'NH', 'New Jersey': 'NJ',
'New Mexico': 'NM', 'New York': 'NY', 'North Carolina': 'NC', 'North Dakota': 'ND', 'Ohio': 'OH', 'Oklahoma': 'OK',
'Oregon': 'OR', 'Pennsylvania': 'PA', 'Rhode Island': 'RI', 'South Carolina': 'SC', 'South Dakota': 'SD',
'Tennessee': 'TN', 'Texas': 'TX', 'Utah': 'UT', 'Vermont': 'VT', 'Virginia': 'VA', 'Washington': 'WA',
'West Virginia': 'WV', 'Wisconsin': 'WI', 'Wyoming': 'WY'}

covid_df['state'] = covid_df['state'].map(us_state_abbrev).fillna(covid_df['state'])
covid_df = covid_df.drop(columns=["fips"])
covid_df.head()
```

```python
covid_df["region"] = covid_df["county"] + ", " + covid_df["state"]
covid_df.head()
```

```python
covid_df.to_csv("please_let_this_one_work.csv", index=False, header=0)
```
