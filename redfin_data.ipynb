```python
import pandas as pd

df = pd.read_csv("county_market_tracker.tsv000", sep="\t", header=0)
df.columns
```

```python
df.columns[df.isna().any()].tolist()
```

```python
df.head()
```

```python
df = df.drop(
    [
        "city", 
        "period_duration", 
        "region_type", 
        "region_type_id", 
        "is_seasonally_adjusted", 
        "last_updated",
        "median_sale_price_mom",
        "median_sale_price_yoy",
        "median_list_price_mom",
        "median_list_price_yoy",
        "median_ppsf_mom",
        "median_ppsf_yoy",
        "median_list_ppsf_mom",
        "median_list_ppsf_yoy",
        "homes_sold_mom",
        "homes_sold_yoy",
        "pending_sales_mom",
        "pending_sales_yoy",
        "new_listings_mom",
        "new_listings_yoy",
        "inventory_mom",
        "inventory_yoy",
        "months_of_supply_mom",
        "months_of_supply_yoy",
        "median_dom_mom",
        "median_dom_yoy",
        "avg_sale_to_list_mom",
        "avg_sale_to_list_yoy",
        "sold_above_list_mom",
        "sold_above_list_yoy",
        "price_drops_mom",
        "price_drops_yoy",
        "off_market_in_two_weeks_mom",
        "off_market_in_two_weeks_yoy",
        "table_id",
        "property_type_id"
    ], 
    axis=1
)
df.head()
```

```python
df = df.dropna()
df.head()
```

```python
df = df.astype(
    {
        "median_sale_price": "int64", 
        "median_list_price": "int64",
        "homes_sold": "int64",
        "pending_sales": "int64",
        "new_listings": "int64",
        "inventory": "int64",
        "median_dom": "int64",
        "parent_metro_region_metro_code": "int64",
        "median_ppsf": "int64",
        "median_list_ppsf": "int64"
    }
)
df.dtypes
```

```python
df = df[df["region"].str.contains("County")]
```

```python
df.to_csv("redfin_county_data.csv", index=False, header=True)
```
